```python
from bs4 import BeautifulSoup
import requests
from openpyxl import Workbook
from utils.product_page import olx_add_information_extractor
import json
import pandas as pd

import time
import csv
import os
```

```python
Data_save_path = 'Scraped_Data.json'
Already_scraped_data_links = []

if os.path.exists(Data_save_path):
  Already_scraped_data = pd.read_json(Data_save_path, lines=True, encoding="utf8")
  Already_scraped_data_links = Already_scraped_data['Product_link'].tolist()

with open(Data_save_path, 'a', encoding="utf8") as f:
    
    for i in range(50):      # Number of pages plus one 
        url = "https://www.olx.com.pk/mobile-phones_c1453?page={}".format(i)
        r = requests.get(url)
        soup = BeautifulSoup(r.content)
        All_items = soup.findAll('li', attrs={'aria-label': 'Listing'})

        for single_item in All_items:
            single_item_link = single_item.find('a')
            
            Product_link = 'https://www.olx.com.pk/'+single_item_link.get('href')
            print(Product_link)
            r = requests.get(Product_link)
            soup2 = BeautifulSoup(r.content, 'html5lib')
            
            if Product_link in Already_scraped_data_links:
              break
            try:
                Scraped_data=olx_add_information_extractor(soup2)
                Scraped_data['Product_link'] = Product_link
            except:
                print('https://www.olx.com.pk/'+single_item_link.get('href'))
                continue
    
            f.write("{0}\n".format(
                json.dumps(Scraped_data, ensure_ascii=False)))
            f.flush()
            time.sleep(2)

     
```
